## Recap

**Markov Chains** are simply sequences of states according to a certain probability matrix of movements.

**Markov Rewards** are Markov Chains with Rewards for each state. We are able to infer the return obtained by following a certain sequence (of states)

**Markov Decision Processes** involve the use of an **agent** and their **policies**. By computing q-values we are able to determine whether performing an action would result in a better policy.

A better policy means a higher Value function. Value functions can be computed in multiple ways. We make informed decisions because we are fully aware of the environment.

**Model-Free** processes involve the use of an **agent**, but we have no information on the environment. For this reason, we must sample episodes to have an *estimate* of the Value Function.

We are not fully aware of the existing environment. We can improve the policy by computing the Q-values, but all the returns will be *estimated* via sampling.

```python
colors = ['red', 'green', 'blue', 'yellow', 'orange', 'purple', 'brown', 'pink']

for episode in episodes:
    time_steps = range(1, len(episode) + 1)
    plt.plot(episode, time_steps, marker='o', linestyle='-', color=colors[0])

plt.xlabel('State')
plt.ylabel('Time Step')
plt.xticks(np.arange(0, 8, step=1))
plt.title('Agent Movement Over Time')
plt.grid(True)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()


for episode in episodes2:
    time_steps = range(1, len(episode) + 1)
    plt.plot(episode, time_steps, marker='o', linestyle='-', color=colors[1])

plt.xlabel('State')
plt.ylabel('Time Step')
plt.xticks(np.arange(0, 8, step=1))
plt.title('Agent Movement Over Time')
plt.grid(True)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```
